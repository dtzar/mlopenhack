```python
from PIL import Image, ImageOps
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

desired_size = 128
im_pth = "C:/Users/nparker/Downloads/gear_images/test.jpeg"

im = Image.open(im_pth)

old_size = im.size  # old_size[0] is in (width, height) format
ratio = float(desired_size)/max(old_size)
new_size = tuple([int(x*ratio) for x in old_size])
im = im.resize(new_size, Image.ANTIALIAS)

fill_color = "#ffffff"

# create a new image and paste the resized on it
new_pil_im = Image.new("RGB", (desired_size, desired_size), fill_color)
new_pil_im.paste(im, ((desired_size-new_size[0])//2,
                (desired_size-new_size[1])//2))

new_pil_im.save(im_pth)
```

```python


simg = Image.open(im_pth)
#simg_gs = simg.convert("L")  # grayscale
#simg_gs.save("C:/Users/nparker/Downloads/gear_images/ImageOps_equalize_sgs.jpg")

# Equalize the image histogram.
dimg = ImageOps.equalize(simg)
dimg.save(im_pth+"1.jpg")
dimg.save("C:/Users/nparker/Downloads/gear_images/ImageOps_equalize_d.jpg")
dimg_gs = dimg.convert("L")  # grayscale
dimg_gs.save("C:/Users/nparker/Downloads/gear_images/ImageOps_equalize_dgs.jpg")

#
# visualize histogram
import matplotlib.pyplot as plt  # https://matplotlib.org
bins = list(range(256))
plt.plot(bins, simg_gs.histogram(), 'r')
plt.plot(bins, dimg_gs.histogram(), 'g')
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.title(im_pth)
plt.legend(
    ('src (grayscale)', 'dest (grayscale)'),
     shadow=True, loc=(0.01, 0.75))
plt.grid(True)
plt.savefig("C:/Users/nparker/Downloads/gear_images/ImageOps_equalize_hist_01.jpg")
```
