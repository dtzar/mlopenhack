```python
def normalize(arr):
    """
    Linear normalization
    http://en.wikipedia.org/wiki/Normalization_%28image_processing%29
    """
    arr = arr.astype('float')
    
    minval = arr.min()
    maxval = arr.max()
    if minval != maxval:
        arr -= minval
        arr *= (255.0/(maxval-minval))
    return arr
```

```python
from PIL import Image, ImageOps
im = Image.open("C:/Users/nparker/Downloads/gear_images/test.jpg")

plt.subplot(121),plt.imshow(im)

arr = np.array(img)
arrnorm = normalize(arr)

new_img = Image.fromarray(normalize(arr).astype('uint8'),'RGB')
new_img.save('C:/Users/nparker/Downloads/gear_images/testnorm.jpg')

#new_img = Image.fromarray(normalize(arr).astype('uint8'),'RGB')
#new_img.save('"C:/Users/nparker/Downloads/gear_images/testnormalised.jpg"')
```
